```python
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.layer import *
```

```python

class ResNet(Layer):
    def __init__(self,out_channels,stride=1):
        super().__init__()
        #Input dimension = output dimension so there's padding ,unlinke valid padding
        first_padding='same'

        if stride !=1:
            first_padding='valid'#It will not get padded ,we'll lose height and width

        self.conv_sequence=Sequential([Conv2D(out_channels,3,stride,padding=first_padding),BatchNormalization(),ReLU(),
        Conv2D(out_channels,3,1,padding='same'),BatchNormalization(),ReLU()])

    def call(self,inputs):
        x=self.conv_sequence(inputs)

        if x.shape == inputs.shape:
            x=x+inputs #this is the skip connection 

        return x
    

layer=ResNet(4)
```
